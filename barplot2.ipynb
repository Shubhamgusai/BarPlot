```python
import matplotlib.pyplot as plt
import pandas as pd
data=pd.read_excel("sde.xlsx")
df=pd.DataFrame(data)
print(df)
plt.bar(df["payment mode"],df["amount"])
plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd
data=pd.read_excel("sde.xlsx")
df=pd.DataFrame(data)
print(df)
grouped_by=df.groupby("payment mode")["amount"].sum()
print(grouped_by)
plt.bar(grouped_by.index,grouped_by.values)
#plt.bar(df["payment mode"],df["amount"])
plt.show()
```
